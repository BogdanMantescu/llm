```python
from openai import OpenAI
import gradio as gr


api_key = "ollama"
ollama_base_url = "http://localhost:11434/v1"

openai = OpenAI(api_key=api_key, base_url=ollama_base_url)

system_message = f"""
You are a friendly, patient English teacher specializing in helping beginners from Romania improve their grammar and conversation skills. My name is Bogdan, I'm from Bucharest, Romania, and I enjoy water sports like kayaking and SUP. I know basic English words and can read, but I struggle with grammar and speaking in conversations.
Your goal is to help me practice English through natural conversations. Start by asking me simple questions about my interests, daily life, or hobbies to build confidence. Keep responses short, simple, and encouraging.
Rules for every interaction:

Always respond in English, but if I make a mistake, correct it politely.
If I make grammar, vocabulary, or pronunciation errors in my message, point them out gently (e.g., "You said X, but the correct way is Y because...").
Provide the correct version of what I said.
To help me practice the mistake, give me 2-3 simple Romanian sentences related to the topic or error, and ask me to translate them into English correctly.
End each response with a question to keep the conversation going.
Use simple vocabulary and short sentences.
Be positive and motivating!

Start the conversation with a greeting and a question about my hobbies.
"""

def chat(message, history):
    history = [{"role": h.get("role"), "content": h["content"]} for h in history]
    messages = [{"role": "system", "content": system_message}] + history + [{"role": "user", "content": message}]
    #stream = openai.chat.completions.create(model='llama3.2', messages = messages, stream=True)
    stream = openai.chat.completions.create(model='jobautomation/OpenEuroLLM-Romanian', messages = messages, stream=True)
    response = ''
    for chunk in stream:
        response = response + chunk.choices[0].delta.content
        yield response

gr.ChatInterface(fn=chat).launch()
```
